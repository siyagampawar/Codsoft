```python
#dropping irrelevant columns
data = churn.drop(['RowNumber','CustomerId','Surname','Gender'],axis = 1)
```

```python
data['Geography'].unique()
```

```python
#encoding data
data1 = pd.get_dummies(data,drop_first = True)
```

```python
data1
```

```python
from sklearn.preprocessing import LabelEncoder
e = LabelEncoder()
data1['Geography_Germany'] = e.fit_transform(data1['Geography_Germany'])
data1['Geography_Spain'] = e.fit_transform(data1['Geography_Spain'])
```

```python
data1.head()
```

```python
data1['Exited'].value_counts()
```

```python
x = data1.drop('Exited',axis = 1)
y = data1['Exited']
```

```python
from sklearn.model_selection import train_test_split
```

```python
#splitting
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42,stratify = y)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train
```

```python
from imblearn.over_sampling import SMOTE
```

```python
x_res,y_res = SMOTE().fit_resample(x,y)
```

```python
y_res.value_counts()
```

```python
x_res = sc.fit_transform(x_res)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
```

```python
rf.fit(x_res,y_res)
```

```python
import joblib
joblib.dump(rf,'churn_model')
```

```python
model = joblib.load('churn_model')
```

```python
model.predict([[619,42,2,0.0,0,0,0,101348.88,0,0]])
```

```python
#accuracy score
from sklearn.metrics import accuracy_score
y_pred = rf.predict(x_test)
accuracy_score(y_test,y_pred)
```
